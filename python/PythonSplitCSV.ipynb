```python
import pandas as pd

# Load the CSV into a dataframe
df = pd.read_csv('rideshare_kaggle.csv')

# Create a list for each table, including the columns necessary for each table
trip_data = ["id", "timestamp", "hour", "day", "month", "datetime", "timezone", "source", "destination", "cab_type", "product_id", "name", "price", "distance", "surge_multiplier"]
location_data = ["id", "longitude", "latitude"]
weather_data = ["id", "temperature", "precipIntensity", "precipProbability", "humidity", "windSpeed", "temperatureHigh", "temperatureHighTime", "temperatureLow", "temperatureLowTime", "sunriseTime", "sunsetTime"]

# Write each list of to seperate CSV files, set index=false to not include index column in CSV
df[trip_data].to_csv("TripData.csv", index=False)
df[location_data].to_csv("LocationData.csv", index=False)
df[weather_data].to_csv("WeatherData.csv", index=False)
```
